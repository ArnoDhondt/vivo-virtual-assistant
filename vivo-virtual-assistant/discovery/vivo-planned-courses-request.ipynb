```python
import requests
from collections import defaultdict


```

```python

url = "https://vivobo-external-production.azurewebsites.net/api/PlannedCourse/FindPlannedCourses"
headers = {'Authorization': '674849c9-3ffa-4d95-b812-2d3da19f0700', 'Content-Type': 'application/json'}

filterBody = {
    "page": {
    "index": 1,
    "size":99999
    }
}

response = requests.post(url, json=filterBody, headers=headers)
```

```python
# we try to retrieve all the courses and add their planned plannedcourses
json_response = response.json()
plannedCourses_list = json_response.get('items')

print("Status Code:", response.status_code)
print("Response JSON:", plannedCourses_list)

course_list = defaultdict(list)

# iterate list
for plannedCourse in plannedCourses_list:
    course_id = plannedCourse["course"]["id"]
    plannedCourse_info = {
        "planned_course_id": plannedCourse["id"],
        "planned_course_title": plannedCourse["title"]
    }

    #append plannedcourse info to courseId
    course_list[course_id].append(plannedCourse_info)

for course, plannedCourses in course_list.items():
    print(f"Course with id: {course} has the following plannedCourses:")
    for plannedCourse in plannedCourses:
        print(f"Id {plannedCourse['planned_course_id']} - {plannedCourse['planned_course_title']}")
```
